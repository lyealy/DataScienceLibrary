```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
%autosave 9999
%matplotlib inline
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2*x -5 + rng.randn(50)
plt.figure(figsize=(12,8))
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(50,1),y)
x_test = np.linspace(0,10,1000)
y_pred = model.predict(x_test.reshape(-1,1))
plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.plot(x_test,y_pred)
```

```python
print('Model slop = {}'.format(model.coef_))
print('Model intercept = {}'.format(model.intercept_))
```

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100,3)
y = 0.5 + np.matmul(X,np.array([1.5,-2,1]).reshape(-1,1))
model.fit(X,y)
print(model.coef_)
print(model.intercept_)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
model = make_pipeline(PolynomialFeatures(7),LinearRegression())
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1*rng.randn(50)
model.fit(x.reshape(-1,1),y)
y_pred = model.predict(x_test.reshape(-1,1))
plt.figure(figsize=(12,8))
plt.scatter(x,y,label='test')
plt.plot(x_test,y_pred,label='predicted line')
plt.legend()
```
